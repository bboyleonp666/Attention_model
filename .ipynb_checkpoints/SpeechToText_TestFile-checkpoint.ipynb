```python
def read_wav(PATH, LABELS, default_rate=16000):
    label_len = len(LABELS)

    # initialize the output array
    wave_count = np.zeros(label_len)
    loss_count = np.zeros(label_len)

    print("LABEL\tTOTAL\tREAD\t<1s COUNT")
    print("-----\t-----\t----\t---------")
    for i, label in enumerate(LABELS):
        files = os.listdir(os.path.join(PATH, label))                                       # list all the files
        waves = [f for f in files if f.endswith('.wav')]                             # get wave files
        wave_len = len(waves)                                                        # get number of wave files
        wave_count[i] = wave_len

        # initialize the temp output array
        tmp_wavData = np.zeros((wave_len, default_rate))
        tmp_wavLabels = np.zeros(wave_len)

        less_than_1s_count = 0
        for j, wav in enumerate(waves):
            path = os.path.join(train_audio_path, label, wav)                        # get path for each wave file
            samples, sample_rate = librosa.load(path, sr=default_rate)               # read file by librosa

            # padding the wave files
            if len(samples) < sample_rate:
                less_than_1s_count += 1                                              # count of files less than 1 second
                white_noise = np.random.normal(0, 0.02, sample_rate-len(samples))    # generate white noise for padding
                samples = np.concatenate((samples, white_noise), axis=None)          # padding the files that is less than 1s

            # output as np.array
            tmp_wavData[j, :] = samples                                              # temporary wave data
            tmp_wavLabels[j] = i                                                     # temporary wave labels

            # print the outcome every ten iterations
            if j == wave_len-1:
                print("{}\t{}\t{}\t{}".format(label, wave_len, j+1, less_than_1s_count), end="\n")
                loss_count[i] = less_than_1s_count
            elif j % 10 == 9:
                print("{}\t{}\t{}\t{}".format(label, wave_len, j+1, less_than_1s_count), end="\r")

        if i == 0:
            wavData = tmp_wavData
            wavLabels = tmp_wavLabels
        else:
            wavData = np.concatenate((wavData, tmp_wavData), axis=0)                 # concatenate info of wave files
            wavLabels = np.concatenate((wavLabels, tmp_wavLabels), axis=None)        # concatenate following labels

    print()
    print("MISSION COMPLETE!!!")
    
    return wavData, wavLabels, wave_count.astype(np.int32), loss_count.astype(np.int32)
```

```python
class MFCC:
    def __init__(self, samples, sample_rate):
        self.samples = samples
        self.sample_rate = sample_rate

    def pre_emphasis(self, alpha=0.95):
        self.samples_emphasized = np.append(self.samples[0], self.samples[1:] - alpha*self.samples[:-1])
        # return self.samples_emphasized
    
    def framing(self, frame_size=0.025, frame_stride=0.01):
        samples_in_frame = int(np.ceil(frame_size*self.sample_rate))                           # number of samples in one frame
        self.total_samples_in_one_frame = samples_in_frame                                     # put it into self
        sample_stride = int(np.ceil(frame_stride*self.sample_rate))                            # sample stride in each iteration
        frame_num = int(np.ceil(
            (len(self.samples_emphasized) - samples_in_frame)/sample_stride) + 1)              # number of iterations

        padding_num = (frame_num-1)*sample_stride + samples_in_frame - len(self.samples_emphasized) # length for padding
        padding = np.zeros(padding_num)                                                             # prepare the padding array
        samples_padded = np.append(self.samples_emphasized, padding)                                # padded sample array

        # index to pick all the overlapping samples
        index_each_frame = np.arange(samples_in_frame)
        index_each_stride = np.linspace(0, len(samples_padded) - samples_in_frame, frame_num).astype(np.int32)
        index = np.tile(index_each_frame, reps=(frame_num, 1)) + np.tile(index_each_stride, reps=(samples_in_frame, 1)).T

        self.frames = np.array([samples_padded[[i]] for i in index])                           # frames is a 2D array
        
    def hamming_window(self):
        # self.frames *= 0.54 - 0.46 * numpy.cos((2 * numpy.pi * n) / (self.total_samples_in_one_frame - 1))
        self.frames *= np.hamming(self.total_samples_in_one_frame)
        
    def stft(self, n_fft=512):
        self.n_fft = n_fft
        magnitude = np.abs(np.fft.rfft(self.frames, n=n_fft))  # magnitude of the FFT
        self.power_spectrum = (1.0/n_fft) * magnitude**2       # power spectrum
    
    def filter_bank(self, n_filter=20):
        low_freq_mel = 0
        high_freq_mel = self.hz2mel(self.sample_rate/2)                               # highest frequency of the Mel
        mel_points = np.linspace(low_freq_mel, high_freq_mel, n_filter+2)             # Equally spaced in Mel scale
        bins = np.floor((self.n_fft+1) * self.mel2hz(mel_points) / self.sample_rate)  # bins for FFT
        
        fbank = np.zeros((n_filter, self.n_fft//2 + 1))
        for j in range(n_filter):
            for i in range(int(bins[j]), int(bins[j+1])):
                fbank[j, i] = (i - bins[j]) / (bins[j+1] - bins[j])
            for i in range(int(bins[j+1]), int(bins[j+2])):
                fbank[j, i] = (bins[j+2] - i) / (bins[j+2] - bins[j+1])
        
        self.mel_fbanks = np.dot(self.power_spectrum, fbank.T)
        self.mel_fbanks = np.where(self.mel_fbanks == 0, np.finfo(float).eps, self.mel_fbanks)
        self.mel_fbanks = 20 * np.log10(self.mel_fbanks)                              # dB
        
    def log_energy(self):
        self.energy = np.log(np.sum(self.mel_fbanks**2, axis=1))

    def hz2mel(self, hz):
        return 2595 * np.log10(1 + hz/700)  # Convert Hz to Mel
    
    def mel2hz(self, mel):
        return 700 * (10**(mel/2595.0) - 1) # Convert Mel to Hz
    
    def default_mfcc(self, nth_derivative=1):
        self.pre_emphasis()
        self.framing()
        self.hamming_window()
        self.stft()
        self.filter_bank()
        self.log_energy()
        
        return np.column_stack((self.energy, self.mel_fbanks))
```

```python
def delta(feat, N):
    """Compute delta features from a feature vector sequence.
    :param feat: A numpy array of size (NUMFRAMES by number of features) containing features. Each row holds 1 feature vector.
    :param N: For each frame, calculate delta features based on preceding and following N frames
    :returns: A numpy array of size (NUMFRAMES by number of features) containing delta features. Each row holds 1 delta feature vector.
    """
    if N < 1:
        raise ValueError('N must be an integer >= 1')
    NUMFRAMES = len(feat)
    denominator = 2 * sum([i**2 for i in range(1, N+1)])
    delta_feat = np.empty_like(feat)
    padded = np.pad(feat, ((N, N), (0, 0)), mode='edge')   # padded version of feat
    for t in range(NUMFRAMES):
        delta_feat[t] = np.dot(np.arange(-N, N+1), padded[t : t+2*N+1]) / denominator   # [t : t+2*N+1] == [(N+t)-N : (N+t)+N+1]
    return delta_feat
```

```python
# the input array of conv1d must be a 3D array so make it into a 3D array
all_wave = np.array(all_wave).reshape(-1, 8000, 1)
```

```python
# split data into training and validation
'''
since we remove some .wav files duration less than 1 second
we use "stratify=y" to split the training data with the ratio of each label
so that to avoid seperating most of one class all into validation
'''

from sklearn.model_selection import train_test_split
x_tr, x_val, y_tr, y_val = train_test_split(
    np.array(all_wave), np.array(y), stratify=y, test_size = 0.2, random_state=777, shuffle=True)
```

```python
## build the model with tensorflow

# input layer
inputs = Input(shape=(8000, 1))

# 1st Conv1D layer
conv = Conv1D(8, 13, padding='valid', activation='relu', strides=1)(inputs)
conv = MaxPooling1D(3)(conv)
conv = Dropout(0.3)(conv)

# 2nd Conv1D layer
conv = Conv1D(16, 11, padding='valid', activation='relu', strides=1)(conv)
conv = MaxPooling1D(3)(conv)
conv = Dropout(0.3)(conv)

# 3rd Conv1D layer
conv = Conv1D(32, 9, padding='valid', activation='relu', strides=1)(conv)
conv = MaxPooling1D(3)(conv)
conv = Dropout(0.3)(conv)

# 4th Conv1D layer
conv = Conv1D(64, 7, padding='valid', activation='relu', strides=1)(conv)
conv = MaxPooling1D(3)(conv)
conv = Dropout(0.3)(conv)

# Flatten layer
conv = Flatten()(conv)

# 1st Dense Layer
conv = Dense(256, activation='relu')(conv)
conv = Dropout(0.3)(conv)

# 2nd Dense Layer
conv = Dense(128, activation='relu')(conv)
conv = Dropout(0.3)(conv)

# output layer
outputs = Dense(len(labels), activation='softmax')(conv)
model = Model(inputs, outputs)
```
